```python
categorical_columns=['protocoltype','service','flag']
```

```python
#One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(X_train[categorical_columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
X_train = pd.concat([X_train, one_hot_df], axis=1)
X_train = X_train.drop(categorical_columns, axis=1)
X_train
```

```python
one_hot_encoded = encoder.transform(X_test[categorical_columns])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
X_test = pd.concat([X_test, one_hot_df], axis=1)
X_test = X_test.drop(categorical_columns, axis=1)
X_test
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
Y_train = to_categorical(Y_train)
```

```python
Y_train
```

```python
model = Sequential()
model.add(Dense(500, activation='relu', input_dim=117))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=20)
```

```python
scores = model.evaluate(X_train, Y_train, verbose=0)
print(scores)
```

```python
pred_test= model.predict(X_test)
pred_test
```
